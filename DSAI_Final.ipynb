```python
# Fırat Atmaca
# 2016105279




import requests


#1

def crawl(url, MaxDept):
    crawled = []
    step = []
    step.append((url, 0))
    index = {}

    while step:
        temp = step.pop(0)
        url = temp[0]
        dept = temp[1]

        if url not in crawled:
            if dept > MaxDept:
                continue

        respones = requests.get(url)

        #print(respones)
        crawled.append(url)
        context = respones.text
        words = context.lower().split()

        for word in words:

            if word in index:
                #print(word)
                new_urls = index[word][0] + [url]



                update_count = index[word][1] + 1
                index[word] = (new_urls, update_count)

            else:
                index[word] = ([url], 1)

        if dept < MaxDept:
            links = get_links(url, context)
            for link in links:
                if link not in crawled:
                    #print(link)
                    step.append((link, dept + 1))

    return index

def get_links(url, context):
    links = []
    href = 'href="'
    start = 0

    while True:
        start = context.find(href, start)
        if start == -1:
            break

        start += len(href)
        end = context.find('"', start)

        link = context[start:end]

        if not link.startswith('http'):
            if link.startswith('/'):

                link = url + link

            else:
                link = url + '/' + link

        links.append(link)
        start = end + 1

    return links




#1 test

test_url = "https://searchengineplaces.com.tr/"
index = crawl(test_url, 2)

for keyword, (urls, count) in index.items():
    print("Keyword:" ,keyword)
    print(count, "result(s) found.")
    print("Urls:")
    for url in urls:
        print(url)
    print()







#1c
def lookupLink(index, url):
    matched_keywords = []
    print(index.items())
    for keyword, (urls, a) in index.items():

        if url in urls:
            matched_keywords.append(keyword)

    keyword_count = len(matched_keywords)

    print("Link: ",url)
    print(keyword_count," keyword(s) found.")


    for keyword in matched_keywords:
        print(keyword)
    print()


# 1c test
test_url = "http://www.searchengineplaces.com.tr/oktayrecommends.html"
lookupLink(index, test_url)


#2
def frequencyOfWord(text):
    text = text.lower()
    text=text.replace(".", "")
    text=text.replace(",", "")
    text=text.replace(";", "")
    text=text.replace(":", "")

    #print(text)

    words=text.split(" ")
    #print(words)

    unique_set ={}
    for word in words:
        if word in unique_set:
            unique_set[word] += 1
        else:
            unique_set[word] = 1

    return unique_set


#2 test
print()
text = "in programming, the hard part isn’t solving problems, but deciding what problems to solve."
print(frequencyOfWord(text))





#3
class Rectangle():
    def __init__(self,len,width):
        self.len =len
        self.width=width

    def Area(self):
        return self.width * self.len


    def diagonal(self):


        result =(self.width**2 + self.len**2)**0.5
        return result



    def __eq__(self, other):

        if self.width==other.width:
            if self.len ==other.len:
               return True
            else:
               return False

        elif self.width == other.len:
            if self.len ==other.width:
                return True
            else:
                return False

        else:
            return False

    def __add__(self,other):
            newWidth = self.width + other.width
            newLen = self.len + other.len

            newRec =Rectangle(newWidth,newLen)

            return newRec

    def __str__(self):

        output ="Rectangle Lenght: "+str(self.len) +" Width: "+str(self.width)
        return output

rec1 = Rectangle(3, 4)
print(rec1.Area())
print(rec1.diagonal())
rec2 = Rectangle(4, 3)
print(rec1 == rec2)
rec3 = rec1 + rec2
print(rec3)
```
